```python
import pandas as pd
import matplotlib.pyplot as plt
 
arquivo = pd.read_csv('PDA_Dados_Cursos_Graduacao_Brasil (1).csv')
 
dados_grafico = arquivo['SITUACAO_CURSO'].value_counts()
 
dados_grafico.plot.pie(autopct='%1.1f%%', startangle=30, figsize=(6, 6))
plt.title('Situação do Curso em Todos os Estados')
 
plt.savefig('situacao.png', bbox_inches='tight')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
 
arquivo = pd.read_csv('PDA_Dados_Cursos_Graduacao_Brasil (1).csv')
 
dados_grafico = arquivo['MODALIDADE'].value_counts()
 
explode = [0.1 + i * 0.1 for i in range(len(dados_grafico))]
 
dados_grafico.plot.pie(autopct='%1.1f%%', startangle=-20, figsize=(8, 8), explode=explode)
 
plt.title('Modalidades Presente em Todos os Estados')
 
plt.legend(dados_grafico.index, loc="best")
 
plt.savefig('modalidade.png', bbox_inches='tight')
plt.show()
```
